```python
import pandas as pd
import math
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon, LineString, MultiPolygon
from shapely.ops import nearest_points
import descartes
import numpy as np
from pyproj import Proj, transform
%matplotlib inline
```

```python
#Asignación de localidad a terreno
cobertura = gpd.read_file('conjunto_de_datos/Cobertura/cobertura.shp')
terrenos = gpd.read_file('conjunto_de_datos/Terrenos/terrenos.shp')
terrenos.head()
```

```python
cobertura.head()
```

```python
pts3 = cobertura.geometry.unary_union
def near(point, pts=pts3):
     # find the nearest point and return the corresponding Clave value
     nearest = cobertura.geometry == nearest_points(point, pts)[1]
     return cobertura[nearest].Clave.get_values()[0]
terrenos['Nearest'] = terrenos.apply(lambda row: near(row.geometry), axis=1)
terrenos.head()
```
